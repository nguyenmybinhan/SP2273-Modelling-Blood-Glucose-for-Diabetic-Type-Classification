## A.1 Introduction & Overview of the paper

**Overview:**
In this paper, a mathematical model for the glucose-insulin regulatory system is developed. A new parameter representing food intake, and hence glucose intake, is introduced. By analyzing the glucose profile of an individual after ingesting a known amount of glucose, we can propose a model for the blood glucose level as a function of time. The parameters of this function can serve as indicators of diabetic status.

## A.2 Literature Review: System of ordinary differential equations (ODE) for Blood-Glucose Dynamics

**1. Some notable definition:**
The glucose disturbance at a compartment is known as the difference between the current glucose level at time $t$ and the basal level in that compartment.

**2. Build-up of the model**:

At time $t$:
- The glucose disturbance of the digestive system (1st compartment) is $D_1(t)$
- Blood-glucose level is $G(t)$
- Effective hormone level (which includes insulin, glucagon, etc) is $H(t)$
- The fasting (basal level) blood glucose is $G_{b}$
- The fasting (basal level) effective hormone levels is $H_b$
- $D_2(t)$: the disturbances of the blood glucose level at time $t$
- $h(t)$: the disturbances of the effective hormone level at time $t$ respectively.
- Thus: $$ D_2(t) = G(t) − G_b$$ and $$h(t) = H(t) − H_b$$

**3. Here are some assumptions:**
1. In **digestive system** **$D_1$**, the rate of decrease in glucose disturbance $\left(\frac{dD_1(t)}{dt}\right)$ is proportional to its glucose disturbance $D_1(t)$ at time $t$.
2. Rate of increase in glucose disturbance in **blood $D_2$** $\left(\frac{dD_2(t)}{dt}\right)$ is proportional to the glucose entering bloodstream from digestive system
$D_1(t)$ at time $t$.
3. Rate of decrease in glucose disturbance in **blood** $D_2$ **not due to hormone** (rate constant: $a_1$) is proportional to its glucose disturbance $D_2(t)$
at time $t$.
4. Rate of decrease in glucose disturbance in **blood** $D_2$ **due to hormone** (rate constant: $a_2$) is proportional to the effective hormone disturbance
$h(t)$ at time $t$.
5. Rate of increase in effective hormone disturbance $\left(\frac{dh(t)}{dt}\right)$ is proportional to the glucose disturbance in blood $D_2(t)$ at time $t$.
6. Rate of decrease in effective hormone disturbance $\left(\frac{dh(t)}{dt}\right)$ is proportional to the effective hormone disturbance $h(t)$ at time $t$.

**4. Our system of ordinary differential equations (ODEs)** under those assumptions:

$$ \frac{dD_1}{dt} = -\frac{1}{\tau} \times D_1(t) $$ 
$$ \frac{dD_2}{dt} = \frac{1}{\tau} \times D_1(t) - a_1 \times D_2 - a_2 \times h$$ 
$$ \frac{dh}{dt} = b_1 \times D_2 + b_2 \times h(t) $$ 



with:
- $\tau$ = Time constant of decreasing glucose level in the digestive system, which is the total time to decrease glucose level to $1/e$
of the maximum value.
- $a_1$ = Rate constant of the hormone independent decrease of glucose level in the blood.
- $a_2$ = Rate constant of the hormone dependent decrease of glucose level in the blood.
- $b_1$ = Rate constant of release of the hormone due to blood glucose disturbance.
- $b_2$ = Rate constant for the removal of the hormone due to disturbance of the blood hormone level.
- $D$ = Amount of food that has been taken.
- $A_G$ = Percentage of glucose obtained in the body from the food that has been taken.

5. The **initial values** of those equations are:

$$ D_1(0) = A_GD = \text {food} = 100 $$ 
$$ D_2(0) = 0$$
$$ h(0) = 0 $$

**6. Solution:**
By solving the sets of 3 ordinary differential equations above, the glucose level in the blood $G(t)$ is:
$$
G(t) = G_b + D_2 = G_b + e^{-\alpha t} \left[ -C_0 \cos(\omega t) + \frac{A_G D + C_0 - \alpha C_0 \tau}{\omega \tau} \sin(\omega t) \right] + C_0 e^{\frac{-t}{\tau}}.
$$

with those parameters:
- $a_1 + b_2 = 2\alpha$
- $ a_1b_2 + a_2b_1 = \omega_{0}^2 $
- $ \omega^2 = \omega_{0}^2 - \alpha^2 $
- $ C_0 = \frac{A_GD(b_2\tau-1)}{(\alpha\tau-1)^2+\omega^2\tau^2}$

## A.3 Limitations in the current model and our inspiration

**Limitations**

**1. Time delay:**

The paper acknowledges its assumptions regarding the pancreas responding instantly to changes in blood glucose and hormones. However, in reality, there is a small time delay in pancreatic response due to these disturbances. Therefore, the paper's modeling is not entirely realistic as it does not incorporate this time delay.

**2. Significance of fit parameters:**
Based on curve fitting to find the best fit parameters of $G_b, \tau, \omega, \alpha, C_0$, the value of the parameter $T_0$, which is a naturally damping of the oscillation curve, is found with the formula $T_0 = \frac{2\pi}{\omega_0}$(hr). The paper finds that if $T_0 \geq 4$, patient can be pre-diabetic or diabetic; and vice versa. While this threshold of $T_0$ can help classify whether a patient is prone to diabetes, it is unable to directly classify the type of diabetes (type 1 or type 2).

**Our inspiration**:

**1. Time delay:**

- We will incorporate a time delay of $\theta$ in our model. This delay specifically accounts for the pancreatic response in the $h(t)$ compartment.

**2. Investigate the main parameters `a1`, `a2`, `b1`, `b2`**:
- We can work backwards to retrieve the main parameters `a1`, `a2`, `b1`, `b2` from the fit parameters $G_b, \omega, \alpha, C_0, \tau$ by solving the following equations:
    - $a_1 + b_2 = 2\alpha$
    - $ a_1b_2 + a_2b_1 = \omega_{0}^2 $
    - $ \omega^2 = \omega_{0}^2 - \alpha^2 $
    - $ C_0 = \frac{A_GD(b_2\tau-1)}{(\alpha\tau-1)^2+\omega^2\tau^2}$
- This is because the main parameters `a1`, `a2`, `b1`, `b2` are direct scientific indicators of whether a person is prone to diabetes and which type of diabetes he is most likely being prone to. For example, if a person has type II diabetes, he is still able to produce insulin yet he is resistant to insulin; in other words, his rate of glucose removed by insulin will be low. Thus, his `a2` parameter is expected to be low.

# B. Our approach: Euler's method for numerical approximation

**Our aim**:

We aim to improve the current model by adding a delay term in the hormone compartment `h(t)` to make the model more realistic. In doing so, we will be solving for main parameters `a1`, `a2`, `b1`, `b2` to investigate whether their respective statistical ranges would suggest the diabetic likelihood of a patient. 

In details, we want to know whether a patient is normal or prone to type II diabetes by studying the statistical distribution of main parameters, specifically `a2` since `a2` denotes the rate of glucose removed by insulin. 
- If `a2` is small enough, we can be certain that a person is resistant to insulin, thus, prone to type II diabetes.
- Otherwise, a person is normal.

**Our approach & the new model of delay differential equations (DDEs):**

Now, we will add a time delay term $ \theta $ in the hormone compartment. The new system will be:
$$ \frac{dD_1(t)}{dt} = -\frac{1}{\tau} \times D_1(t) $$ 
$$ \frac{dD_2(t)}{dt} = \frac{1}{\tau} \times D_1(t) - a_1 \times D_2(t) - a_2 \times h(t-\theta)$$ 
$$ \frac{dh(t)}{dt} = b_1 \times D_2(t) + b_2 \times h(t-\theta) $$ 

**Our approach**:

Since the above system does not contain standard delay differential equations, we decided not to solve it mathematically. We will use Euler's method for numerical solution for the ease of computation. Moreover, Euler's method also satisfies our aim in investigating the main parameters `a1`, `a2`, `b1`, `b2` since we are able to tune those parameters for fitting our model.

**Euler's method**:

Based on small approximation:
$$\frac{dD_1(t)}{dt} \approx \frac{\Delta {D_1}}{\Delta t} = -\frac{1}{\tau} \times D_1(t)$$
Hence: $$\Delta {D_1} \approx -\frac{1}{\tau} \times D_1(t) \times \Delta t $$
To calculate new $D_1$: $$D_{1}(t) \approx \Delta {D_1} + D_{1}(t-1)$$

Similarly: $$D_{2}(t) \approx \Delta {D_2} + D_{2}(t-1)$$ and $$ h(t) \approx \Delta {h} + h(t-1)$$

For better visualisation, the following table can help us keep track of the changes in each compartment. Here, we choose $ \Delta t = 0.01 $.

|:--:|:--:|:--:|:--:|:--:|:--:|:--:|:--:|

# I. Curve fitting to get fitting parameters `Gb`, `alpha`, `omega`, `tau`, `C0`

The paper did not explicitly demonstrate how curve-fitting was done to retrieve the parameters `Gb`, `alpha`, `omega`, `tau`, `C0`. Thus, we will use the mathematical function suggested in the paper to perform our own curve-fitting of a given blood-glucose profile of a patient. 

Note that the blood-glucose level $G(t)$ is a function of time $t$ as follows:
$$
G(t) = G_b + D_2 = G_b + e^{-\alpha t} \left[ -C_0 \cos(\omega t) + \frac{A_G D + C_0 - \alpha C_0 \tau}{\omega \tau} \sin(\omega t) \right] + C_0 e^{\frac{-t}{\tau}}.
$$

For any given blood-glucose profile, if we know the amount of food $A_GD$, we can perform curve-fitting to find the fit parameters $G_b, \omega, \tau, C_{0}, \alpha $. We will do so using `lmfit` function in Python, which provides a high-level interface to non-linear optimization and curve fitting problems for Python.

The function `fit_glucose_levels` takes in our blood-glucose profile in `time` and `experimental_data` as input. The expected data type of `time` and `experimental_data` here is an array. 
- Inside the function `fit_glucose_levels`, we define another `glucose_in_blood` function which takes `params` as a variable to store our fit parameters (`Gb`, `alpha`, `omega`, `tau`, `C0`), with `time` and `experimental_data`. It returns the error between our predicted blood-glucose level based on the best-fit parameters `Gb`, `alpha`, `omega`, `tau`, `C0` based on our mathematical function to that of experimental data. 
- Next, we specify the range of our fit parameters `Gb`, `alpha`, `omega`, `tau`, `C0`. Currently, the range of fit parameters are decided based on the known fit values from the papers.
- Then, we perform the fit with `minimize` function that minimizes the error of prediction between our own prediction to that of experimental data. It takes the following inputs:
    - Our `glucose_in_blood`: since it returns the error of prediction
    - `params`: that contains our fit parameters name and specified range
    - `args=`: our arguments in a tuple which will be `time` and `experimental_data` since these are arguments of the `fit_glucose_levels` function
    - `method = 'nelder'`
    - All the fitting parameters are stored in the variable `fitted_params`
- After that, we retrieve our parameters with `fitted_params.params[<< name_of_variable>>].value` and round it to 2 digits

Using our datasets, we will specify $A_GD = 100 = food$ in our curve-fitting model.

```python
# uncomment the below line to install lmfit
# pip install lmfit 
```

```python
#import necessary packages
import numpy as np
import pandas as pd
from lmfit import Parameters, minimize, fit_report
import matplotlib.pyplot as plt
import os 
import random
```

```python
def fit_glucose_levels(time, experimental_data):
    time = np.array(time)
    
    def glucose_in_blood(params, time, experimental_data):
        
        #specify our parameters
        Gb = params['Gb']
        alpha = params['alpha']
        omega = params['omega']
        tau = params['tau']
        C0 = params['C0']
        
        #mathematical function
        glucose_fitted = Gb + np.exp(-alpha*time) * (-C0 * np.cos(omega*time) + ((100 + C0 - alpha*C0*tau) / (omega*tau)) * np.sin(omega*time)) + C0 * np.exp(-time / tau)
        
        #return error of prediction as the difference between fitted glucose and experimental data
        return glucose_fitted - experimental_data

    # Create lmfit Parameters
    params = Parameters()
    params.add('Gb', min=70, max=100)
    params.add('alpha', min=0, max=2)
    params.add('omega', min=0.5, max=3)
    params.add('tau', min=0.01, max=1)
    params.add('C0', min=100, max=200)

    # Perform the fit
    fitted_params = minimize(glucose_in_blood, params, args=(time, experimental_data,), method='nelder')

    # Retrieve fitted parameter values
    Gb = round(fitted_params.params['Gb'].value,2)
    alpha = round(fitted_params.params['alpha'].value,2)   
    omega = round(fitted_params.params['omega'].value,2)  
    tau = round(fitted_params.params['tau'].value,2)   
    C0 = round(fitted_params.params['C0'].value,2)

    return Gb,alpha,omega,tau,C0
```

# II. Calculate main parameters `a1`, `a2`, `b1`, `b2`

- We can work backwards to retrieve the main parameters `a1`, `a2`, `b1`, `b2` from the fit parameters $G_b, \omega, \alpha, C_0, \tau$ by solving the following equations:
    - $a_1 + b_2 = 2\alpha$
    - $ a_1b_2 + a_2b_1 = \omega_{0}^2 $
    - $ \omega^2 = \omega_{0}^2 - \alpha^2 $
    - $ C_0 = \frac{A_GD(b_2\tau-1)}{(\alpha\tau-1)^2+\omega^2\tau^2}$

However, only `a1`, `b2` are directly solvable. Based on the knowledge of $G_b, \omega, \alpha, C_0, \tau$, and `a1`, `b2`, we only can get the product of `a2` and `b1`, stored in the variable `a2_b1`. Thus, we have to randomise `a2`. `b1` will be the division of `a2_b1`/`a2`.

```python
def get_parameters(AgD, Gb, alpha, omega, C0, tau):
    b2 = (C0 * ((alpha * tau - 1) ** 2 + omega ** 2 * tau ** 2) / AgD + 1) / tau
    a1 = 2 * alpha - b2
    a2_b1 = omega ** 2 + alpha ** 2 - a1 * b2
    a2 = random.uniform(0, a2_b1) #very important the range!
    b1 = a2_b1 / a2
    
    # Round each parameter to two decimal places using map and lambda
    a1, a2, b1, b2 = map(lambda x: round(x, 2), (a1, a2, b1, b2))
    
    return a1, a2, b1, b2
```

# III. Our approach using Euler's method in ODEs

We test our approach in using Euler's method in the system of ODEs like that suggested in the paper.

We define a function `get_glucose_profile_with_no_time_lag` that takes in `food`, `time`, `experimental_data`, and `delta_t` as inputs. 
- `food`: amount of glucose a person takes in at time `t` = 0
- `time`: points of time measured in glucose-tolerance test stored as a list
- `experimental_data`: the amount of glucose measured at each respective point of time stored as a list
- `delta_t`: our step size of increment in time. We choose $\Delta t = 0.01$ in all our code.

The function returns us list of `t` and `g`:
- `t`: list of points of time from `0` to maximum of `time` of the glucose-tolerance test
- `g`: list of our modelled blood-glucose levels at each respective point of time `t` 

```python
def get_glucose_profile_with_no_time_lag(food, time, experimental_data, delta_t): 
    #fit the exponential curve to get our fitted parameters
    time_array = np.array(time)
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(time_array, experimental_data)

    #retrieve our main parameters of differential equations
    a1,a2,b1,b2 = get_parameters(food, Gb, alpha, omega, C0, tau)
    
    #initial value of t
    t_val = 0
    #list to store time
    t = [t_val]

    #initial value of D1
    d1_initial = food
    d1_val = d1_initial
    
    #list to store d1
    d1 = [d1_val]

    #intial value of d2
    d2_initial = 0
    d2_val = d2_initial
    #list to store d2
    d2 = [d2_val]
    
    #initial value of h
    h_initial = 0
    h_val = h_initial
    h = [h_val]

    #initial value of glucose in blood
    g = [Gb]

    #setting cut off index for loop 
    loop_cut_off = int(time[-1]/delta_t)
    
    for i in range(loop_cut_off):
        #compartment 1: D1
        delta_d1 = -1/tau * d1_val * delta_t
        d1_val = d1_val + delta_d1
        d1.append(d1_val)
    
        #compartment 2: D2
        delta_d2 = (1/tau * d1_val - a1*d2_val - a2*h_val)*delta_t
        d2_val = d2_val + delta_d2
        d2.append(d2_val)
    
        #compartment 3: h
        delta_h = (b1*d2_val - b2*h_val)*delta_t
        h_val = h_val + delta_h
        h.append(h_val)
        
        #glucose in blood
        g_val = Gb + d2_val
        g.append(g_val)
        
        #increment time
        t_val += delta_t
        t.append(t_val)
    return t,g
```

Now, we test our function `get_glucose_profile_with_no_time_lag(food, time, experimental_data, delta_t)`. We will use experimental data of person A for testing.

```python
t_no_delay, g_no_delay = get_glucose_profile_with_no_time_lag(100, t_A, G_A, 0.01)

#convert our list of t_no_delay to array to perform the G(t) function
t_no_delay = np.array(t_no_delay)

# our fit parameters: Gb_A,alpha_A,omega_A,tau_A,C0_A
g_by_math = Gb_A + np.exp(-alpha_A*t_no_delay) * (-C0_A * np.cos(omega_A*t_no_delay) + ((100 + C0_A - alpha_A*C0_A*tau_A) / (omega_A*tau_A)) * np.sin(omega_A*t_no_delay)) + C0_A * np.exp(-t_no_delay / tau_A)

#plot the curve
plt.plot(t_no_delay, g_no_delay, label = 'No delay')
plt.plot(t_no_delay, g_by_math, label = 'G(t) function')
plt.scatter(t_A, G_A, color = 'r', label = 'Experimental data')
plt.legend()
plt.xlabel('time $t$/ h')
plt.ylabel('blood glucose level/ unit')
plt.title('Euler\'s method for numerical approximation with NO TIME DELAY')
plt.show()
```

# IV. Euler's approach in delay differential equations (DDE) with `time_delay` ($ \theta $)

## 1. Convert time delay (in min) to theta (in hour)

Time delay is usually in minutes. However, since our `t` is in hour, we need to convert time delay in minute to hour. This is done by taking the time delay dividing by 60.

This is done by the function `convert_min_to_hour(time_lag)` which takes in the variable `time_lag` in minute and returns a time lag in hour.

```python
def convert_min_to_hour(time_lag):
    return time_lag/60
```

## 2. Applying our delay differential equations system

Recall our new system with time lag:
$$ \frac{dD_1(t)}{dt} = -\frac{1}{\tau} \times D_1(t) $$ 
$$ \frac{dD_2(t)}{dt} = \frac{1}{\tau} \times D_1(t) - a_1 \times D_2(t) - a_2 \times h(t-\theta)$$ 
$$ \frac{dh(t)}{dt} = b_1 \times D_2(t) + b_2 \times h(t-\theta) $$ 

Since our hormone disturbance function $h(t)$ only accepts non-negative input, for any time `t` $\in [0, \theta]$, the input of our function $h(t-\theta)$ is negative as $-\theta \leq t - \theta \leq 0$. Thus, any non-negative inputs will make function `h` zero. In other words:
$$ \forall t \in [0, \theta]: h(t-\theta) = 0  $$ 

Thus, for any $ 0 \leq t < \theta$, our system is written as:
$$ \frac{dD_1(t)}{dt} = -\frac{1}{\tau} \times D_1(t) $$ 
$$ \frac{dD_2(t)}{dt} = \frac{1}{\tau} \times D_1(t) - a_1 \times D_2(t) $$ 
$$ \frac{dh(t)}{dt} = b_1 \times D_2(t) $$ 

And for any $ t \geq \theta$, our system will be:
$$ \frac{dD_1(t)}{dt} = -\frac{1}{\tau} \times D_1(t) $$ 
$$ \frac{dD_2(t)}{dt} = \frac{1}{\tau} \times D_1(t) - a_1 \times D_2(t) - a_2 \times h(t-\theta)$$ 
$$ \frac{dh(t)}{dt} = b_1 \times D_2(t) + b_2 \times h(t-\theta) $$ 

## 3. Our function `get_glucose_profile_with_time_lag`

The function takes the following variables as inputs:
- `food`: amount of glucose taken in at time `t=0`. We use `100` as this is the amount that the paper takes in
- `time`: list of points of time the blood-glucose levels are sampled from a person after he takes in a known amount of glucose
- `experimental_data`: list of sampled blood-glucose levels at each respective point of time
- `delta_t`: increment step size in our `t_val`
- `person_label`: to label person A or B

**How does this function work?**:
- Based on the sampled glucose-levels in a range of times (`time` and `experimental_data`), we first perform curve_fitting to retrieve fit parameters `Gb`,`alpha`,`omega`,`tau`,`C0`
- From the fit parameters `Gb`,`alpha`,`omega`,`tau`,`C0`, calculate main parameters `a1`,`a2`,`b1`,`b2`
- Convert `time_lag` in minutes to hour
- Find the `loop_cut_off` by taking the largest value of `time` dividing by `delta_t`
- Find the `breaking_point` where time starts exceed time delay (`t_val` > `theta`):
    - Before the `breaking_point` (`i` < `breaking_point`): the pancreas is still taking time to respond to changes in blood-glucose level.
        - Use `try`-`except` to catch `IndexError`, which occurs as `i - breaking_point` becomes negative, which leads to an attempt to access an index that is out of the range of the list h.
        - When `IndexError` occurs, the `except` code runs and we remove `a2 * h[i-breaking_point]` and `b2 * h[i-breaking_point]` in compartment D2 and h respectively
    - After the `breaking_point` (`i` $\geq$ `breaking_point`):
        - No`IndexError` occurs, and thus the `except` code does not run
- Set initial values & change at each step `i` of the loop:
    - For time :
        - Initial value is 0
        - Variable `t_val` keeps incrementing with step size of `delta_t`
        - Variable `t` is a list to store different values of `t_val`
    - For comparment D1:
        - Initial value is `food`, the amount of glucose taken in at time `t_val = 0`
        - Step size `delta_d1` = $-\frac{1}{\tau}$ `* d1_val * delta_t` (since $\Delta D_1 \approx -\frac{1}{\tau} \times D_1(t) \times \Delta t $)
        - Variable `d1_val` keeps incrementing with step size of `delta_d1`
    - For compartment d2:
        - Initial value is `0`
        - Step size `delta_d2` = ($\frac{1}{\tau}$ `* d1_val - a1 * d2_val - a_2 * h[i-breaking_point]) * delta_t` (since $\Delta D_2 \approx [\frac{1}{\tau} \times D_1(t) - a_1 \times D_2(t) - a_2 \times h(t-\theta)] \times \Delta t$
        - Variable `d1_val` keeps incrementing with step size of `delta_d1`
    - For compartment hormone:
        - Initial value is `0`
        - Step size `delta_h` = `(b1 * times d2_val + b_2 * h[i-breaking_point])* delta_t` (since $ \Delta h \approx [b_1 \times D_2(t) + b_2 \times h(t-\theta)] \times \Delta t $)
        - Variable `h_val` keeps incrementing with step size of `delta_h`
    - For blood glucose:
        - Intital value is at basal level `Gb`
        - Keeps incrementing with step size of `d2_val` because $G(t) = G_b + D_2(t)$

```python
def get_glucose_profile_with_time_lag(food, time, experimental_data, time_lag, delta_t, person_label): 
    
    #fit the exponential curve to get our fitted parameters
    time_array = np.array(time)
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(time_array, experimental_data)

    #retrieve our main parameters of differential equations
    a1,a2,b1,b2 = get_parameters(food, Gb, alpha, omega, C0, tau)

    #Convert time_lag in minutes to hour
    theta = convert_min_to_hour(time_lag)

    #now, set intial values of t, D1, D2, h, g
    #initial value of t
    t_val = 0
    #list to store time
    t = [t_val]

    #initial value of D1
    d1_initial = food
    d1_val = d1_initial
    
    #list to store d1
    d1 = [d1_val]

    #intial value of d2
    d2_initial = 0
    d2_val = d2_initial
    #list to store d2
    d2 = [d2_val]
    
    #initial value of h
    h_initial = 0
    h_val = h_initial
    h = [h_val]

    #initial value of glucose in blood
    g = [Gb]

    breaking_point = 0
    loop_cut_off = int(time[-1]/delta_t)
    testing_t = 0
   
    #find the breaking_point where time starts >= theta
    for i in range(loop_cut_off):
        if testing_t < theta:
            testing_t += delta_t
            breaking_point += 1
    
    for i in range(loop_cut_off):     
        #compartment d1
        delta_d1 = -(1/tau)*d1_val*delta_t
        d1_val += delta_d1
        d1.append(d1_val)

        #compartment d2
        if i >= breaking_point:
            delta_d2 = (1/tau * d1_val - a1*d2_val - a2 * h[i-breaking_point])*delta_t
        else:
            delta_d2 = (1/tau * d1_val - a1*d2_val)*delta_t
        d2_val += delta_d2
        d2.append(d2_val)

        #compartment h
        if i >= breaking_point:
            delta_h = (b1*d2_val - b2 * h[i-breaking_point])*delta_t
        else:
            delta_h = (b1*d2_val)*delta_t
        h_val += delta_h
        h.append(h_val)

        #blood glucose
        g_val = Gb + d2_val
        g.append(g_val)

        #increase time
        t_val += delta_t
        t_val = round(t_val,2)
        t.append(t_val)
    
    #calculate G(t) using mathematical function
    math = []
    for i in t:
        mathematical_G = Gb + np.exp(-alpha*i)*(-C0*np.cos(omega*i)+ (food+C0-alpha*C0*tau)/(omega*tau)*np.sin(omega*i)) +C0*np.exp(-i/tau)
        math.append(mathematical_G)

    #get data for model if there is no time lag
    t_no_lag, g_no_lag = get_glucose_profile_with_no_time_lag(food, time, experimental_data, delta_t)

    return t,math, g_no_lag, g, time_lag, time, experimental_data, theta, Gb, person_label, a2
```

## 7. Finding optimal `time_delay`

Now, we will find the optimal `time_delay` for each person
- The optimal `time_delay` is the value of `time_delay` that yields minimum error between the modelled glucose level and the actual experimental glucose level
- We choose Root-mean-square deviation (RMSE) as our error metric because it has the ability to penalise large deviations

Formula for RMSE:

 $\text{RMSE} = \sqrt{ \sum_{i = 1}^{n} \frac{( \widehat{y}_i - y_i)^2}{n} }$

The function `best_delay` will perform the finding of optimal `time_delay`. It takes `time_range`, `food`, `time`, `experimental_data`, `delta_t`, `person_label` as inputs and outputs the optimal the dictionary that has a key as `time_delay` and value as its corresponding RMSE of modelling using that `time_delay`.

```python
def best_delay(time_range, food, time, experimental_data, delta_t, person_label):
    error_list = {}
    for time_delay in time_range:
        t,_, _, g, *_, a2 = get_glucose_profile_with_time_lag(food, time, experimental_data, time_delay, delta_t, person_label)
        d_pred = dict(zip(t,g)) #key = t, g = blood glucose
        d_actual = dict(zip(time, experimental_data))
        mse_val = 0
        
        for i in time:
            predicted_glucose = d_pred[i]
            experimental_glucose = d_actual[i]
            mse = abs(predicted_glucose - experimental_glucose)**2
            mse_val += mse
        rmse = (mse_val/len(time))**0.5
        error_list[time_delay] = rmse
    return error_list
```

## 8. At `time_delay` = 5, how different `a2` affect the modelling curve?

Undestanding that we have randomised value of parameter `a2` in the function `get_parameters(AgD, Gb, alpha, omega, C0, tau)`, randomising `a2` may affect our modelling curve. 

Thus, we want to provide a visualisation of how the modelling curve will differ with different values of `a2` while keeping the `time_delay`= 5min as this is the optimal `time_delay` regardless of value of `a2`.

We rewrite a function that calculates only `a1`, `b2`, and product of `a2`, `b1` as `a2_b1`. This is because those values can be calculated explicitly from fit parameters `Gb`, `alpha`, `omega`, `C0`, `tau`.

```python
def get_a1_b2(food, Gb, alpha, omega, C0, tau):
    b2 = (C0 * ((alpha * tau - 1) ** 2 + omega ** 2 * tau ** 2) / food + 1) / tau
    a1 = 2 * alpha - b2
    a2_b1 = omega ** 2 + alpha ** 2 - a1 * b2
    
    # Round each parameter to two decimal places 
    a1, b2, a2_b1 = map(lambda x: round(x, 2), (a1, b2, a2_b1))
    
    return a1, b2, a2_b1
```

Next, assume that we have values of all main parameters `a1`, `a2`, `b1`, `b2`, we write the function `get_glucose_profile_using_ab_params` to perform Euler's method and returns the blood-glucose modelling profile as `t` and `g`, together with `time_lag`. 

```python
def get_glucose_profile_using_ab_params(a1,a2,b1,b2, tau, time_lag, food, Gb, time, delta_t):
    #setting initial values and variables 
    theta = convert_min_to_hour(time_lag)
    
    #initial value of t
    t_val = 0
    #list to store time
    t = [t_val]

    #initial value of D1
    d1_initial = food
    d1_val = d1_initial
    
    #list to store d1
    d1 = [d1_val]

    #intial value of d2
    d2_initial = 0
    d2_val = d2_initial
    #list to store d2
    d2 = [d2_val]
    
    #initial value of h
    h_initial = 0
    h_val = h_initial
    h = [h_val]

    #initial value of glucose in blood
    g = [Gb]

    breaking_point = 0
    loop_cut_off = int(time[-1]/delta_t)
    testing_t = 0
   
    #find the breaking_point where time starts >= theta
    for i in range(loop_cut_off):
        if testing_t < theta:
            testing_t += delta_t
            breaking_point += 1
            
    for i in range(loop_cut_off):
        #compartment d1
        delta_d1 = -(1/tau)*d1_val*delta_t
        d1_val += delta_d1
        d1.append(d1_val)

        #compartment d2
        if i >= breaking_point:
            delta_d2 = (1/tau * d1_val - a1*d2_val - a2 * h[i-breaking_point])*delta_t
        else:
            delta_d2 = (1/tau * d1_val - a1*d2_val)*delta_t
        d2_val += delta_d2
        d2.append(d2_val)

        #compartment h
        if i >= breaking_point:
            delta_h = (b1*d2_val - b2 * h[i-breaking_point])*delta_t
        else:
            delta_h = (b1*d2_val)*delta_t
        h_val += delta_h
        h.append(h_val)

        #blood glucose
        g_val = Gb + d2_val
        g.append(g_val)

        #increase time
        t_val += delta_t
        t_val = round(t_val,2)
        t.append(t_val)
    
    return t, g, time_lag
```

## 9. Find optimal parameter `a2` for each patient (with `time_delay` = 5min)

Now, we want to find the optimal value of parameter `a2` that best fits the experimental data. This is done by our function `find_optimal_a2` that takes in `food`, `time`, `experimental_data`, `time_lag`, `delta_t`, `person_label` as inputs, similarly to other functions. 

However, in this function, we will:
1. Retrieve the range of `a2` based on its product with `b1`. The range of `a2` will be: 0 < `a2` < `a2_b1`
2. Loop through each value of `a2` and find the RMSE associated with the main parameters
3. Put in a dictionary with key as `a2`, and value as its associated `RMSE`

```python
def find_optimal_a2(food, time, experimental_data, time_lag, delta_t, person_label):
    
    #fit the exponential curve to get our fitted parameters
    time_array = np.array(time)
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(time_array, experimental_data)

    #retrieve a1 & b2 & product of (a2 * b1)
    a1,b2, a2_b1 = get_a1_b2(food, Gb, alpha, omega, C0, tau)

    a2 = np.arange(0.01, a2_b1, 0.01)
    a2_choice = list(map(lambda x: round (x,2), list(a2)))

    actual_record = dict(zip(time, experimental_data)) #key = time, value = experimental glucose G
    error_record = {} #key = time_lag, value = RMSE

    for a2_val in a2_choice:
        #get parameters
        a2 = a2_val
        b1 = round(a2_b1/a2_val)
        
        #get prediction
        t, g, _ = get_glucose_profile_using_ab_params(a1,a2,b1,b2, tau, time_lag, food, Gb, time, delta_t)
        predict_record = dict(zip(t,g)) #key = time, value = predicted glucose G

        #mse
        mse = 0
        #go to each actual data point and calculate MSE
        for time_val in time:
            actual_G = actual_record[time_val]
            predict_G = predict_record[time_val]
            mse += (actual_G - predict_G)**2
        
        #calculate rmse
        rmse = (mse/len(time))**0.5

        #append to error record
        error_record[a2_val] = rmse
    return error_record
            
        
        
```

## 10. Compare error of modelling between mathematical function & Euler's method

```python
def Euler_with_no_time_lag(food, time, experimental_data, delta_t, time_delay): 
    #fit the exponential curve to get our fitted parameters
    time_array = np.array(time)
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(time_array, experimental_data)

    time = np.array(time)
    math = Gb + np.exp(-alpha*time) * (-C0 * np.cos(omega*time) + ((100 + C0 - alpha*C0*tau) / (omega*tau)) * np.sin(omega*time)) + C0* np.exp(-time / tau)
    math_model = dict(zip(time, math))
    
    #find best fit a2
    a2_dict = find_optimal_a2(100, time, experimental_data, time_delay, 0.01, '')
    a2 = min(a2_dict.items(), key = lambda x: x[1])[0]

    #calculate main parameters a1, b2
    a1, b2, a2_b1 = get_a1_b2(food, Gb, alpha, omega, C0, tau)
    b1 = a2_b1/a2
    
    #initial value of t
    t_val = 0
    #list to store time
    t = [t_val]

    #initial value of D1
    d1_initial = food
    d1_val = d1_initial
    
    #list to store d1
    d1 = [d1_val]

    #intial value of d2
    d2_initial = 0
    d2_val = d2_initial
    #list to store d2
    d2 = [d2_val]
    
    #initial value of h
    h_initial = 0
    h_val = h_initial
    h = [h_val]

    #initial value of glucose in blood
    g = [Gb]

    #setting cut off index for loop 
    loop_cut_off = int(time[-1]/delta_t)
    
    for i in range(loop_cut_off):
        #compartment 1: D1
        delta_d1 = -1/tau * d1_val * delta_t
        d1_val = d1_val + delta_d1
        d1.append(d1_val)
    
        #compartment 2: D2
        delta_d2 = (1/tau * d1_val - a1*d2_val - a2*h_val)*delta_t
        d2_val = d2_val + delta_d2
        d2.append(d2_val)
    
        #compartment 3: h
        delta_h = (b1*d2_val - b2*h_val)*delta_t
        h_val = h_val + delta_h
        h.append(h_val)
        
        #glucose in blood
        g_val = Gb + d2_val
        g.append(g_val)
        
        #increment time
        t_val += delta_t
        t.append(t_val)
    return t,g
```

```python
def calculate_rmse(food, time, experimental_data, delta_t, time_delay):

    #actual experimental data
    time = list(map(lambda x: round(x,2),time))
    actual_data = dict(zip(time, experimental_data)) #key:time value: experimental_data
    
    #model by math
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(np.array(time), experimental_data)
    time = np.array(time)
    math = Gb + np.exp(-alpha*time) * (-C0 * np.cos(omega*time) + ((100 + C0 - alpha*C0*tau) / (omega*tau)) * np.sin(omega*time)) + C0* np.exp(-time / tau)
    math_model = dict(zip(time, math))
    
    #find best fit a2
    a2_dict = find_optimal_a2(100, time, experimental_data, time_delay, 0.01, '')
    a2 = min(a2_dict.items(), key = lambda x: x[1])[0]

    #calculate main parameters a1, b2
    a1, b2, a2_b1 = get_a1_b2(food, Gb, alpha, omega, C0, tau)
    b1 = a2_b1/a2

    #Euler's with delay
    t_delay, g_delay, _ = get_glucose_profile_using_ab_params(a1,a2,b1,b2, tau, time_delay, food, Gb, time, delta_t)
    t_delay = list(map(lambda x: round(x,2),t_delay))
    delay = dict(zip(t_delay, g_delay))

    #Euler's with no delay
    t_no_delay, g_no_delay = Euler_with_no_time_lag(food, time, experimental_data, delta_t, time_delay)
    t_no_delay = list(map(lambda x: round(x,2),t_no_delay))
    no_delay = dict(zip(t_no_delay, g_no_delay))

    #variable to store rmse of each case
    rmse_by_math = 0
    rmse_by_euler_no_delay = 0
    rmse_by_euler_delay = 0

    #traverse through the time to get error at each point of experimental data
    for t_val in time:
        model_by_math = math_model[t_val]
        model_by_euler_no_lag = no_delay[t_val]
        model_by_euler_lag = delay[t_val]
        actual_glucose = actual_data[t_val]
        
        rmse_by_math += (actual_glucose - model_by_math)**2
        rmse_by_euler_no_delay += (actual_glucose - model_by_euler_no_lag)**2
        rmse_by_euler_delay += (actual_glucose - model_by_euler_lag)**2
    
    rmse_by_math = (rmse_by_math/len(time))**0.5
    rmse_by_euler_no_delay = (rmse_by_euler_no_delay/len(time))**0.5
    rmse_by_euler_delay = (rmse_by_euler_delay/len(time))**0.5

    return f'With time delay: {time_delay}min, Error modelling by math: {rmse_by_math:.2f} Error modelling by Euler with no delay {rmse_by_euler_no_delay:.2f}. Error modelling by Euler with delay {rmse_by_euler_delay:.2f}'
```

```python
def plotting_Euler_best_fit(food, time, experimental_data, delta_t, person_label, time_delay):
    Gb,alpha,omega,tau,C0 = fit_glucose_levels(np.array(time), experimental_data)
    
    #find best fit a2
    a2_dict = find_optimal_a2(100, time, experimental_data, time_delay, 0.01, '')
    a2 = min(a2_dict.items(), key = lambda x: x[1])[0]

    #calculate main parameters a1, b2
    a1, b2, a2_b1 = get_a1_b2(food, Gb, alpha, omega, C0, tau)
    b1 = a2_b1/a2
    
    #Euler's with delay
    t_delay, g_delay, _ = get_glucose_profile_using_ab_params(a1,a2,b1,b2, tau, 5, food, Gb, time, delta_t)
    t_delay = list(map(lambda x: round(x,2),t_delay))

    #Euler's with no delay
    t_no_delay, g_no_delay = Euler_with_no_time_lag(food, time, experimental_data, delta_t, time_delay)
    
    #math
    t = np.arange(0,time[-1],0.01)
    math = Gb + np.exp(-alpha*t) * (-C0 * np.cos(omega*t) + ((100 + C0 - alpha*C0*tau) / (omega*tau)) * np.sin(omega*t)) + C0* np.exp(-t / tau)
    
    plt.scatter(time, experimental_data, c = 'red', label = 'experimental data')
    plt.plot(t_delay, g_delay, label =f'Euler with {time_delay} delay')
    plt.plot(t, math, label = 'by math')
    plt.plot(t_no_delay, g_no_delay, label ='Euler with no delay')
    plt.legend()
    plt.title(f'Comparison of modelling between different methods of person {person_label} with a2 :{a2}')
    plt.show()
    #plt.close()
    
    
```

# E. Limitations & Significance of our work

**1. Improved the current model to make it more realistic**
- By incorporating time delay into the current model, we are able to mirror real-life blood-glucose regulation dynamics more realistically. However, our approach are still unable to model as accurately as the mathematical function since mathematical function is the absolute solution to the system.
- Thus, our limitation still lies on finding optimal parameter individually instead of simultaneously although the latter can be more computationally expensive yet accurate. 

**2. Studied the main parameters & the potential use of them to predict the diabetic status**
- While the current model use ODEs to study the fit parameters (`Gb`, `alpha`, `omega`, `tau`, `C0`), we delve deeper into the main parameters `a1`, `a2`, `b1`, `b2`. The main parameters are more significant and could be demonstrate a more insightful understanding of the blood-glucose regulation system in each person since they are direct scientific indicators of the blood-glucose dynamics in our body.
- Moreover, as we study the best-fit of parameters `a2` and `time_delay` via their statistical distribution, this can be the foundation for our future work in classifing patient's diabetic status based on patient's main parameters.
- In that sense, we can determine whether a person is prone to type II diabetes by checking if their main parameters `a2` are significantly low enough to ensure that they are resistant to insulin and confirm their type II diabetic status.
- Similarly, we can predict whether a person is prone to type I diabetes by checking if their main parameters `b1` (rate of hormone released in response to changes in blood-glucose level) is significantly low enough to ensure they are struggling to produce insulin and confirm their type I diabetic status.
